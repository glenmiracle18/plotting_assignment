Histogram of CO2 distribution

```python
carbon_dioxide = df['co2']
plt.figure(figsize=(10,8))
plt.hist(carbon_dioxide, bins=30, edgecolor='black')
plt.title('Carbon Dioxide Distribution')
plt.yticks(np.arange(0, 450, 50))
plt.xlabel('Carbon Dioxide')
plt.ylabel('Frequency')
# plt.yscale('log')
plt.show()
```

```python
# Create histogram
plt.figure(figsize=(10, 8))
# loop through each unique vclass and plot for each
for vclass in df['VClass'].unique():
    plt.hist(df[df['VClass'] == vclass]['displ'], bins=10, edgecolor='black', alpha=0.5, label=vclass)
plt.xlabel('Fuel Disposal')
plt.ylabel('Frequency')
plt.title('Fuel Efficiency by Vehicle Class')
plt.legend()
plt.show()
```

```python
# Carbon emission vs engine displacement
plt.hist(df['comb'], edgecolor='black', bins=30)
plt.xlabel('Combustion (C)')
plt.ylabel('Frequench')
plt.title('Combustion distribuiton')
plt.show()
```

```python
# function to remove all the non numerical columns
def remove_colums(df):
    """
    return dataset with no non numerical value
    :param df: input dataFrame
    :return: clean dataFrame
    """
    return df.select_dtypes(include=["number"])

data = remove_colums(df)
data.head()
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, vmin=-1, vmax=1, cmap='vlag')
plt.title('Correlation Heat Map')

plt.show()
```
